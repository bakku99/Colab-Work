```python
import os

# üîç Walk through Colab directory to find EBGaramond OTFs
def find_ebgaramond_font_dir(base_path='/content'):
    for root, dirs, files in os.walk(base_path):
        for file in files:
            if file.lower().startswith('ebgaramond') and file.lower().endswith('.otf'):
                return root
    return None

# Detect font directory
font_dir = find_ebgaramond_font_dir()

if font_dir:
    print(f"‚úÖ EB Garamond font directory found: {font_dir}")
else:
    print("‚ùå EB Garamond fonts not found!")
```

```python
# üö® Compile once and stop on first error
import subprocess

tex_file_path = '/content/drive/MyDrive/Malkoros/DM Advice/Crafting 2/Getting Perfect.tex'  # <-- UPDATE THIS PATH

compile_command = [
    "xelatex",
    "-interaction=errorstopmode",  # Stop on error immediately
    "-file-line-error",            # Show file and line number in errors
    tex_file_path
]

# First pass
result = subprocess.run(compile_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

# Output compile log
print(result.stdout)

# Raise error if compile failed
if result.returncode != 0:
    print(result.stderr)
    raise RuntimeError("‚ùå LaTeX compilation failed!")
else:
    print("‚úÖ First compilation pass complete.")
```

```python
# üöÄ Compile twice nonstop for TOC and references
import subprocess

tex_file_path = '/content/drive/MyDrive/Malkoros/DM Advice/Crafting 2/FixHeaderAndFooter.tex'  # <-- UPDATE THIS PATH

compile_command = [
    "xelatex",
    "-interaction=nonstopmode",    # Don't stop on errors
    "-file-line-error",            # Show file and line number in errors
    tex_file_path
]

# First pass
print("üöÄ Starting first compilation pass...")
first_pass = subprocess.run(compile_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
print(first_pass.stdout)

# Second pass
print("üöÄ Starting second compilation pass...")
second_pass = subprocess.run(compile_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
print(second_pass.stdout)

# Check if second pass failed
if second_pass.returncode != 0:
    print(second_pass.stderr)
    raise RuntimeError("‚ùå LaTeX compilation failed on second pass!")
else:
    print("‚úÖ Second compilation pass complete. Document should be fully updated!")
```
